# Modeliszation of the string

## Physical model of plucked string

When plucked by a finger or a plectrum, the string is moved away from its initial equilibriumposition. During the stick phase, a contact exists between the exciter and the string. When the restoring force due to tension (see *Fig. 1*) becomes slightly higher than the frictional force, the string slips under the finger.

A simplified model for the starting transient of the string is presented below. It is assumed that the string leaves the exciter without initial velocity at string position $x_0$ with the initial profile (see *Fig. 2*):
\begin{align*}
y(0,t) = \begin{cases}
    \frac{hx}{x_0} & 0 \leq x \leq x_0,\\
    \frac{h(L-x)}{L-x_0} & x_0 \leq x \leq L,
\end{cases}
\end{align*}

Starting from $\rho(x)S(x)\ddot{y}(x,t) - \frac{\partial}{\partial x}\left(T(x)\frac{\partial y}{\partial x}\right) = f(x,t)$, where $\rho(x)$ is the mass per unit length, $S(x)$ is the cross-sectional area, $T(x)$ is the tension, and $f(x,t)$ is the external force, the modal method involves searching for solutions of $y(x,t)=\sum_{n} \phi_n(x)q_n(t)$. This equation must be satisfied in particular at time $t=0$, which leads to
\begin{align*}
q_n(0) &= \frac{1}{m_n}\int_{0}^{L}\rho(x)S(x)y(0,t)\phi_n(x)dx \\
\dot{q}_n(0) &= \frac{1}{m_n}\int_{0}^{L}\rho(x)S(x)\dot{y}(0,t)\phi_n(x)dx
\end{align*}

Since the initial string velocity is taken equal to zero, we get $\dot{q}_n(0)=0$, and the remaining unknown variables of the problem are the initial generalized
displacements $q_n(0)$. Using the orthogonality properties of the modes, we find
\begin{equation*}
q_n(0)=\frac{1}{m_n}\int_{0}^{L}\rho S y(0,t)\phi_n(x)dx = \frac{1}{m_n}\int_{0}^{L}\rho S y(0,t)\sin(k_nx)dx
\end{equation*}
from which we derive
\begin{equation*}
q_n(0)=\frac{2hL^2}{n^2\pi^2 x_0 (L-x_0)}\sin(k_n x_0)
\end{equation*}

The expression for $q_n(t)$ is provided by the equation of an oscillator with no applied force:
\begin{equation*}
\ddot{q}_n(t) + \omega_n^2 q_n(t) = 0
\end{equation*}
which, given the initial displacement and velocity conditions, leads to $q_n(t) = q_n(0)\cos(\omega_n t)$. 

In summary, limit conditions at $x=0$ and $x=L$ is given by $ y(0,t) = y(L,t) = 0 $, which leads to $\phi_n(x) = \sin(k_n x)$ where $k_n = \frac{n\pi}{L}$. Initial conditions are given by $\dot{y}(x,0) = 0$, which leads to $q_n(0) = \frac{2hL^2}{n^2\pi^2 x_0 (L-x_0)}\sin(k_n x_0)$. Therefore, the general expression of the transverse displacement of the string is given by:
\begin{equation*}
y(x,t) = \sum_{n=1}^{\infty} \frac{2hL^2}{n^2\pi^2 x_0 (L-x_0)}\sin(k_n x_0)\sin(k_n x)\cos(\omega_n t)
\end{equation*}

## Add stiffness to the model

In the ideal vibrating string, the only restoring force for transverse displacement comes from the string tension; specifically, the transverse restoring force is equal the net transverse component of the axial string tension. Consider in place of the ideal string a bundle of ideal strings, such as a stranded cable. When the cable is bent, there is now a new restoring force arising from some of the fibers being compressed and others being stretched by the bending. This force sums with that due to string tension. Thus, stiffness in a vibrating string introduces a new restoring force proportional to bending angle. It is important to note that string stiffness is a linear phenomenon resulting from the finite diameter of the string.

In typical treatments, bending stiffness adds a new term to the wave equation that is proportional to the fourth spatial derivative of string displacement:
\begin{equation*}
\frac{1}{c_t^2}\frac{\partial^2 y}{\partial t^2} = \frac{\partial^2 y}{\partial x^2} - \frac{1}{\kappa^2}\frac{\partial^4 y}{\partial x^4}
\end{equation*}
where $c_t = \sqrt{\frac{T}{\rho_L}}$ is the wave speed, $T$ is tension, $\rho_L$ is the linear density, and $\kappa = \sqrt{\frac{T}{EI}}$ is the stiffness parameter, $T$ is the tension, $E$ is the Young's modulus, and $I$ is the moment of inertia of the string. Theoretical proof of this equation can be found in [this paper](https://courses.physics.illinois.edu/phys406/sp2017/Lecture_Notes/Waves/PDF_FIles/Waves_2.pdf).

Under the limit condition of null displacement, we get
\begin{equation*}
\omega_n = \frac{n\pi c_t}{L}\sqrt{1 + \left(\frac{n\pi}{\kappa L}\right)^2}
\end{equation*}
and $y(x,t)=\sum_{n=1}^{\infty} b_n \sin(k_n x)\cos(\omega_n t)$ where $b_n = \frac{2F}{TL} \frac{\sin(k_n x_0)}{k_n^2} \frac{1}{1 + \left(\frac{k_n}{\kappa}\right)^2}$, $F$ is the force applied to the string, and $T$ is the tension of the string.

## Add damping to the model

The factor of quality $Q(f)$ decides the damping of the string. For the low frequencies, the friction force from the air $Q_{air}$ is the dominant damping factor. For the high frequencies, the viscoelasticity of the string $Q_{visc}$ is the dominant damping factor. We give these factors directly by
\begin{align*}
Q^{-1}(f) &= Q_{air}^{-1}(f) + Q_{visc}^{-1}(f) + Q_{th}^{-1}(f) \\
Q_{air}^{-1}(f) &= \frac{R}{2\pi \rho_L f} \\
Q_{visc}^{-1}(f) &= \frac{4 \pi^2 \rho_L EIf^2 \delta_{ve}}{T^2} \\
Q_{th}^{-1}(f) &= \text{constant}
\end{align*}
where
\begin{align*}
R=\pi^2 \rho_{air} f \frac{d^2}{2}\left(\frac{\sqrt{2}}{M}+\frac{1}{2M^2}\right) \text{ with } 
\begin{cases}
   M = \frac{d}{4}\sqrt{\frac{2\pi f}{\mu_a}}\\
    \mu_a = \frac{\eta}{\rho_{air}}
\end{cases}
\end{align*}
is resistance, $\eta$ is the viscosity of the air, $\rho_{air}$ is the density of the air, $\rho_L$ is the linear density, $f$ is the frequency, $T$ is the tension, $E$ is the Young's modulus, $d$ is the diameter of the string, and $\delta_{ve}$ is the viscoelasticity of the string.

The damping term is written as $e^{-\frac{\pi f_n}{Q(f_n)}t}$, where $f_n$ is the frequency of the $n$-th mode. The general expression of the transverse displacement of the string is given by:
\begin{equation*}
y(x,t) = \sum_{n=1}^{\infty} b_n \sin(k_n x)\cos(\omega_n t)e^{-\frac{\pi f_n}{Q(f_n)}t}
\end{equation*}
where 
\begin{equation*}
f_n = \frac{n}{2L} \sqrt{\frac{T}{\rho_L}} \sqrt{1 + B n^2} \text{ with } B = \left(\frac{\pi}{\kappa L}\right)^2
\end{equation*}

## Simulation for the model

### imports

```python
import numpy as np
from IPython.display import Audio
import matplotlib.pyplot as plt
```

### Parameters

```python
config = {
    'sample_rate': 44100, # Hz
    'duration': 5, # seconds
    'number of modes': 5,
    'longer': 0.648, # m
    'initial displacement': 0.02, # m
    'percentage of position of excitation': 0.2, # %
    'percentage of position of sensor': 0.6, # %
    'density of material': 7850, # kg/m^3
    'density of air': 1.2, # kg/m^3
    'viscoealstic damping': 1.8e-5, # Pa.S
    'diameter of string': 0.3e-3, # m
    'tension': 100, # N
    'exciter force': 0.05, # N
    'Young modulus': 2.1e11, # Pa

}
```

```python
n_mode = config['number of modes']
sample_rate = config['sample_rate']
time_duration = config['duration']
L = config['longer']
h = config['initial displacement']
x0 = L * config['percentage of position of excitation']
xc = L * config['percentage of position of sensor']
diameter = config['diameter of string']
rho_L = config['density of material'] * np.pi * diameter**2 / 4
rho_air = config['density of air']
T_corde = config['tension']
c_corde = np.sqrt(T_corde / rho_L)
I = np.pi * diameter**4 / 64 # moment of inertia
kappa = np.sqrt(T_corde / (config['Young modulus'] * I))
F = config['exciter force']
E = config['Young modulus']
eta = config['viscoealstic damping']

x_scale = np.linspace(0, L, 1000)
t_scale = np.linspace(0, time_duration, time_duration * sample_rate)

def get_k_n(n, L):
    return n * np.pi / L

def get_omega_n(n, L, c_corde):
    return get_k_n(n, L) * c_corde

def get_q_n_0(n, L, h, x0):
    return (2 * h * L**2) / (n**2 * np.pi**2 * x0 * (L - x0)) * np.sin(get_k_n(n, L) * x0)

```

### calculate different modes

```python
def calulate_mode(n, L, x_scale=x_scale):
    return np.sin(get_k_n(n, L) * x_scale)

# plot modes
modes = [calulate_mode(n_model, L) for n_model in range(1, n_mode + 1)]
fig, ax = plt.subplots()
fig.set_size_inches(10, 6)
for mode in modes:
    ax.plot(x_scale, mode)
ax.set_title('Modes')
plt.grid()
plt.xlabel('x (m)')
plt.ylabel('displacement (m)')
plt.legend([f'n = {n}' for n in range(1, n_mode + 1)])
plt.show()
```

```python
def calculate_position(n_max, L, h, x0, t, x_scale):
    displacement =  0
    for n in range(1, n_max + 1):
        displacement += get_q_n_0(n, L, h, x0) * np.cos(get_omega_n(n, L, c_corde) * t) * calulate_mode(n, L, x_scale)
    return displacement

# plot position
time_steps = np.arange(0, 0.5, 0.1)
displacements = [calculate_position(n_mode, L, h, x0, t, x_scale) for t in time_steps]
fig, ax = plt.subplots()
fig.set_size_inches(10, 6)
for disp in displacements:
    ax.plot(x_scale, disp)
ax.set_title('Position')
plt.grid()
plt.xlabel('x (m)')
plt.ylabel('displacement (m)')
plt.legend([f't = {t}' for t in time_steps])
plt.show()
```

```python
# generate sound at sensor position

def generate_sound(n_max, L, h, x0, xc, sample_rate, time_duration):
    sound = np.zeros(int(sample_rate * time_duration))
    for i in range(len(sound)):
        sound[i] = calculate_position(n_max, L, h, x0, i / sample_rate, xc)
    return sound

sound = generate_sound(n_mode, L, h, x0, xc, sample_rate, 1)

fig, ax = plt.subplots()
fig.set_size_inches(10, 6)
ax.plot(sound)
ax.set_title('Sound')
plt.grid()
plt.show()

Audio(sound, rate=sample_rate)
```

### add stiffness

```python
def get_stif_omega_n(n, L, c_corde, kappa=kappa):
    return n * np.pi / L * c_corde * np.sqrt(1 + (n * np.pi / (kappa * L))**2)

def get_b_n(L, F, kn, x0):
    return 2 * F * np.sin(kn * x0) / (T_corde * L * kn**2 * (1 + (kn / kappa)**2))

def generate_stif_sound(n_max, L, x0, xc, t_scale):
    sound = np.zeros(len(t_scale))
    for i in range(len(t_scale)):
        sound[i] = 0
        for n in range(1, n_max + 1):
            kn = get_k_n(n, L)
            bn = get_b_n(L, F, kn, x0)
            omega_n = get_stif_omega_n(n, L, c_corde)
            sound[i] += bn * np.cos(omega_n * t_scale[i]) * np.sin(kn * xc)
    return sound

fn = [get_stif_omega_n(n, L, c_corde)/ (2 * np.pi) for n in range(1, n_mode + 1)]

fig, ax = plt.subplots()
fig.set_size_inches(10, 6)
ax.plot(range(n_mode),fn)
ax.scatter(range(n_mode), fn)
ax.set_title('Frequencies')
plt.grid()
plt.xlabel('mode')
plt.ylabel('f (Hz)')
plt.show()
    
sound = generate_stif_sound(n_mode, L, x0, xc, t_scale)

fig, ax = plt.subplots()
fig.set_size_inches(10, 6)
ax.plot(t_scale, sound)
ax.set_title('Sound')
plt.grid()
plt.show()

Audio(sound, rate=sample_rate)
```

### add damping

```python
def get_f_n(n, L, T_corde, rho_L, E, I):
    B = np.pi**2 * E * I / (T_corde * L**2)
    return n / (2 * L) * np.sqrt(T_corde / rho_L) * np.sqrt(1 + B * n**2)

def get_M(f, d, eta):
    mu_a = eta / rho_air
    return d/4 * np.sqrt(2 * np.pi * f / mu_a)

def calculate_Q(f, d, eta, rho_L, T_corde, E, I, delta_ve=3.5e-3, Q_th=7e3):
    M = get_M(f, d, eta)
    R = np.pi**2 * rho_air * f * d**2 / 2 * (np.sqrt(2)/M + 1 / (2 * M**2))
    Q_air_inv = R / (2 * np.pi * f * rho_L)
    Q_visc_inv = 4 * np.pi**2 * rho_L * E * I * f**2 * delta_ve / (T_corde**2)
    Q_th_inv = 1 / Q_th
    return 1 / (Q_air_inv + Q_visc_inv + Q_th_inv)

def generate_damping_sound(n_max, L, F, x0, xc, c_corde, rho_L, E, I, t_scale):
    sound = np.zeros(len(t_scale))
    for i in range(len(t_scale)):
        sound[i] = 0
        for n in range(1, n_max + 1):
            kn = get_k_n(n, L)
            bn = get_b_n(L, F, kn, x0)
            omega_n = get_stif_omega_n(n, L, c_corde)
            fn = get_f_n(n, L, T_corde, rho_L, E, I)
            sound[i] += bn * np.cos(omega_n * t_scale[i]) * np.sin(kn * xc) * np.exp(- np.pi * fn / calculate_Q(fn, diameter, eta, rho_L, T_corde, E, I) * t_scale[i])
    return sound

fn = [get_f_n(n, L, T_corde, rho_L, E, I) for n in range(1, n_mode + 1)]

fig, ax = plt.subplots()
fig.set_size_inches(10, 6)
ax.plot(range(n_mode),fn)
ax.scatter(range(n_mode), fn)
ax.set_title('Frequencies')
plt.grid()
plt.xlabel('mode')
plt.ylabel('f (Hz)')
plt.show()

f_scale = np.linspace(10, 10000, 10000)
Q = [calculate_Q(f, diameter, eta, rho_L, T_corde, E, I) for f in f_scale]

fig, ax = plt.subplots()
fig.set_size_inches(10, 6)
ax.semilogx(f_scale, Q)
ax.set_title('Q')
plt.grid()
plt.xlabel('f (Hz)')
plt.ylabel('Q')
plt.show()

sound = generate_damping_sound(n_mode, L, F, x0, xc, c_corde, rho_L, E, I, t_scale)

fig, ax = plt.subplots()
fig.set_size_inches(10, 6)
ax.plot(t_scale, sound)
ax.set_title('Sound')
plt.grid()
plt.show()

Audio(sound, rate=sample_rate)
```
